### **What is `print()`?**
When you write code, Python is doing a lot of work quietly in the background.
`print()` is how you make Python speak — it shows you what’s happening *right now* on the screen.

You’ll use `print()` when you want to:

* see a message
* check the value of a variable
* understand your logic step-by-step (debugging)

One important thing to keep straight: **`print()` is not `return`.**

* `print()` **shows** something to the screen (mainly for you).
* `return` **sends** a value back from a function (so your program can use it later).

That’s the core idea: **`print()` is for visibility, `return` is for results.**

## 4) Escape sequences

Escape sequences start with a backslash `\` and give special meaning.

Common ones:

| Escape Characters | Name          | Use in Code |
| ----------------- | ------------- | ----------- |
| `"`               | Double quotes | `\"`        |
| `'`               | Single quotes | `\'`        |
| `\`               | Backslash     | `\\`        |
| `\n`              | New line      | `\n`        |
| `\b`              | Backspace     | `\b`        |
| `\t`              | Tab           | `\t`        |

#### \\" Double quotes

````python
print("Hello \"World\"")
````

#### \\' Single Quotes

````python
print('Hello \'World\'')
````

#### \ backslash

````python
print("This is double backslash \\\\")
````

## 5) Printing multiple values (comma behavior)

You can print many things in one `print()` call.
By default, `print()` puts a **space** between items.


````python
print("Anand",25,98.5)
````

## 6) `sep` — change the separator

By default:
```python
print("A", "B", "C")   # A B C
```

Use `sep` to change the thing placed between printed items.


````python
print("A", "B", "C", sep="-")
````

````python
print("2026", "01", "30", sep="/")
````

````python
print("apple", "banana", "mango", sep=", ")
````

## 7) `end` — control what happens at the end

By default, `print()` ends with a new line (`\n`).

Use `end` to:
- keep output on the same line
- add custom endings (like `!`, `...`, etc.)


````python
print("Loading", end="...")
print("done!")
````

````python
print("A", end=" ")
print("B", end=" ")
print("C")
````

## 8) Blank lines, spacing, and clean formatting

````python
# You can also create spacing with \n
print("Top\n\nBottom")
````

### 10) Calculations in Python (using print())

Python can do math directly inside print().

You write the expression, and Python prints the result.

````python
print("Addition        (+) : ", 7 + 2)
print("Subtraction     (-) : ", 7 - 2)
print("Multiplication  (*) : ", 7 * 2)
print("Float Division  (/) : ", 7 / 2)
print("Integer Division(//): ", 7 // 2)
print("Modulus         (%) : ", 7 % 2)
print("Exponent        (**): ", 7 ** 2)
````

### 11) Data Types
Python has different “kinds of values” — these are called **data types**.
Once you know the type, you can usually predict what operations will work and what errors you might get.

### 1) Built-in types (Python gives you these)

These are the ones you’ll use every day:

* **Numbers:** `int`, `float`, `complex` *(and `bool` is a special numeric type: `True/False`)*
* **Text:** `str`
* **Collections (multiple values):**

  * Sequences Ordered collections: `list`, `tuple`, `range`
  * Unique items: `set`, `frozenset`
  * Key–value pairs: `dict`
* **Binary data:** `bytes`, `bytearray`, `memoryview` *(used in files, images, networking, etc.)*
* **“Nothing” value:** `None` *(type is `NoneType`)*

### 2) User-defined types (you create these)

When built-ins aren’t enough, you make your own type using a **class**.
That’s how you model real things like `Student`, `BankAccount`, `Car`, etc.

#### Numbers: int, float, complex and bool

````python
print(6)
print(type(6))
# big integer (Python int can be very large)
print(10**100)
print(type(10**100))
````

````python
# float
print(3.14)
print(type(3.14))

# scientific notation is float
print(1e3)
print(type(1e3))

# very large float becomes "inf" (infinity)
print(1e309)
print(type(1e309))
````

````python
# complex
print(2 + 3j)
print(type(2 + 3j))
````

````python
# boolean (bool)
print(True)
print(False)
print(type(True))
````

#### Text

````python
# string (str)
print("Hello Python")
print(type("Hello Python"))
````

#### Container Types

````python
# list
print([1, 2, 3])
print(type([1, 2, 3]))
````

````python
# tuple
print((1, 2, 3))
print(type((1, 2, 3)))
````

````python
# set (unique values)
print({1, 2, 2, 3})
print(type({1, 2, 2, 3}))
````

````python
# dictionary (key-value)
print({"name": "Anand", "role": "Developer"})
print(type({"name": "Anand", "role": "Developer"}))

# range (sequence generator)
print(range(5))
print(type(range(5)))

# None (means "no value")
print(None)
print(type(None))
````

#### Binary data:

````python
# bytes (binary data)
print(b"ABC")
print(type(b"ABC"))

# bytearray (mutable bytes)
print(bytearray(b"ABC"))
print(type(bytearray(b"ABC")))

# memoryview (view into bytes-like data)
print(memoryview(b"ABC"))
print(type(memoryview(b"ABC")))
````

````python
# user-defined type (class)
class Student:
    pass

s = Student()
print(s)
print(type(s))
````
