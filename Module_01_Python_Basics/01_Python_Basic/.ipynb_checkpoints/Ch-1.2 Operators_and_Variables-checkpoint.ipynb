#### 1. Arithmetic operators

```python
print("Addition(+)        = ",5+4)

print("Subtraction(-)     = ",5-4)

print("Multipliction(*)   = ",5*4)

print("division(/)        = ",5/2)

print("Integer divison(//)= ",5//2)

print("Modulus(%)         = ",5%2)  # reminder output
```

#### 2. Comparison/Relation operators

```python
print("Greater than(>)              =",4>5)

print("Less than(<)                 =",4<5)

print("Greater than or equal to(>=) =",4>=4)

print("Less then or equal to(<=)    =",4<=4)

print("Equal(==)                    =",4==4)

print("Not Equal(!=)                =",4!=4)
```

#### 3. Logical operators

```python
# And
True + False  = False
True + True   = True
False + False = False

# or
True + False = True
True + True = True

# not

False = True
```

```python
print("and = ",0 and 1)

print("or  = ",0 or 1)

print("not = ",not 1)
```

#### 4. Bitwise Operators (operators work binary number)

```python
# bitwise and # 2 = 010 and 3 = 110

print("bitwise and(&) = ",2 & 3)

# bitwise or
print("bitwise or(|)  = ",2 | 3)

# bitwise Xor
print("bitwise Xor(^) = ",2 ^ 3)        # if both same then 0 ,if both different then 1

# bitwise not
print("bitwise not(~) = ",~ 3)    # This operatord use in image processing or computer vision
```

## What are the main data types in Python, and how do you determine the data type of a variable?
1. **Numeric types:** These include integers (int), floating-point numbers (float), and complex numbers (complex).

2. **Sequence types:** These include lists (list), tuples (tuple), and ranges (range).

3. **Text type:** This is the string type (str).

4. **Mapping type:** This is the dictionary type (dict).

5. **Set types:** These include sets (set) and frozen sets (frozenset).

6. **Boolean type:** This is the Boolean type (bool).

`# to determine the data type of  a vaiable we use type() `

```python
a = 23
b = "Hello"
c = {"Hello":23,"23":"Hello"}
d = [1,2,3,4]
e = 23.5
f = True

print(f'The Datatype of Variable a:- {type(a)}')
print(f'The Datatype of Variable b:- {type(b)}')
print(f'The Datatype of Variable c:- {type(c)}')
print(f'The Datatype of Variable d:- {type(d)}')
print(f'The Datatype of Variable e:- {type(e)}')
print(f'The Datatype of Variable f:- {type(f)}')
```

# String Methods
1. len()    `function`

3. upper()  `method`
4. title()  `method`
5. count() `method`

```python
# 1. len  function
Name = "Harry Potter"

lenght = len(Name)

print(lenght)
```

```python
# 2. lower()  `method`
Name = "Harry Potter"

print(Name.lower())
```

```python
# 3. upper()  `method`
Name = "Harry Potter"

print(Name.upper())
```
