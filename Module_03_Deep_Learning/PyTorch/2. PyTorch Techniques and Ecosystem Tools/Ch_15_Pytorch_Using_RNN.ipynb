```python
import re
```

```python
def tokenize(text):
    if isinstance(text, str): 
        text = text.lower()
        
        # We added \d to the brackets to catch all numbers
        # Alternatively, you could write 0-9 inside the brackets
        text = re.sub(r'[!-@#$%^&*()_+><:;?\'\d]', '', text)
        
        return text.split()
    return text
```

```python
tokenize(df['Question'][4])
```

```python
# Vocab (for unknow words)

vocab = {'<UNK>':0}
```

```python
def build_vocab(row):
    # if pd.isna(row['Question']) or pd.isna(row['Answer']):
    #     return

    tokenized_question = tokenize(row['Question'])
    tokenized_answer = tokenize(row['Answer'])

    for token in tokenized_question + tokenized_answer:
        if token not in vocab:
            vocab[token] = len(vocab)
```

```python
df.apply(build_vocab,axis=1)
```

```python
def text_to_indices(text, vocab):

    return [
        vocab.get(token, vocab['<UNK>'])
        for token in tokenize(text)
    ]
```

```python
import torch
from torch.utils.data import DataLoader , Dataset
```

```python
class QADataset(Dataset):

    def __init__(self,df,vocab):
        self.df = df
        self.vocab = vocab

    def __len__(self):
        return self.df.shape[0]

    def __getitem__(self,index):

        numerical_question = text_to_indices(self.df.iloc[index]['Question'],self.vocab)
        numerical_answer = text_to_indices(self.df.iloc[index]['Answer'],self.vocab)

        return torch.tensor(numerical_question) , torch.tensor(numerical_answer)
```

```python
dataset = QADataset(df,vocab) 
```

```python
dataloader = DataLoader(dataset,batch_size=1,shuffle=True)
```

```python
for question , answer in dataloader:
    print(question,answer[0])
```
