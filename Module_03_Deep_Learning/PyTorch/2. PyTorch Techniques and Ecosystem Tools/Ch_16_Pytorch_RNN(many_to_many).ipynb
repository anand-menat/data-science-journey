```python
import re
```

```python
def tokenize(text):
    if isinstance(text, str): 
        text = text.lower()
        
        # We added \d to the brackets to catch all numbers
        # Alternatively, you could write 0-9 inside the brackets
        text = re.sub(r'[!-@#$%^&*()_+><:;?\'\d]', '', text)
        
        return text.split()
    return text
```

```python
tokenize(df['Question'][0])
```

```python
# Vocab

vocab = {'<PAD>':0,'SOS':1,'EOS':2,'<UNK>':3}
```

```python
def build_vocab(row):

    tokenized_question = tokenize(row['Question'])
    tokenized_answer = tokenize(row['Answer'])

    for token in tokenized_question + tokenized_answer:
        if token not in vocab:
            vocab[token] = len(vocab)
```

```python
def text_to_indices(text, vocab,add_eos=False):

    indices = [
        vocab.get(token,vocab['<UNK>'])
        for token in tokenize(text)
    ]
    if add_eos:
        indices.append(vocab['EOS'])
    return indices
```

```python
import torch
from torch.utils.data import DataLoader , Dataset
```

```python
class QADataset(Dataset):

    def __init__(self,df,vocab):
        self.df = df
        self.vocab = vocab

    def __len__(self):
        return self.df.shape[0]

    def __getitem__(self,index):

        numerical_question = text_to_indices(self.df.iloc[index]['Question'],self.vocab)
        numerical_answer = text_to_indices(self.df.iloc[index]['Answer'],self.vocab,add_eos=True)

        return torch.tensor(numerical_question) , torch.tensor(numerical_answer)
```

```python
dataset = QADataset(df,vocab) 
```

```python
dataloader = DataLoader(dataset,batch_size=1,shuffle=True)
```

```python
# Model Building 
# Simple RNN

class Seq2SeqRNN(nn.Module):
    def __init__(self,vocab_size,embed_dim=50,hidden_dim = 64):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size,embed_dim)
        self.encoder = nn.RNN(embed_dim,hidden_dim,batch_first=True)
        self.decoder = nn.RNN(embed_dim,hidden_dim,batch_first=True)
        self.fc = nn.Linear(hidden_dim,vocab_size)

    def forward(self,question,answer=None,max_len=10):

        # Encoder
        embedded_q = self.embedding(question)
        _ , hidden = self.encoder(embedded_q)

        # Decoder - <SOS> token
        batch_size = question.size(0)
        decoder_input = torch.ones(batch_size , 1 , dtype=torch.long)

        outputs = []
        seq_len = answer.size(1) if answer is not None else max_len

        for i in range(seq_len):
            embedded_d = self.embedding(decoder_input)
            out , hidden = self.decoder(embedded_d,hidden)
            prediction = self.fc(out.squeeze(1))  # [1,vocab_size]
            outputs.append(prediction)

            if answer is not None:
                decoder_input = answer[:,i].unsqueeze(1)  # teacher forcing
            else:
                decoder_input = prediction.argmax(dim=1).unsqueeze(1)
        return torch.stack(outputs,dim=1)   # [1,seq_len , vocab_size]
```

```python
learning_rate = 0.001
epochs = 100
```

```python
model = Seq2SeqRNN(len(vocab))
```

```python
# loss , optimizer 

criterion = nn.CrossEntropyLoss(ignore_index=0)
optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)
```

```python
# Training Loop


for epoch in range(epochs):
    total_loss =0 

    for question , answer in dataloader:
        if answer.numel() == 0:
            continue
        optimizer.zero_grad()

        output = model(question,answer)
        output_flat = output.view(-1,len(vocab)) # [1,seq_len,vocab_size]
        target_flat = answer.view(-1)
    
        loss = criterion(output_flat , target_flat)
        loss.backward()

        optimizer.step()

        total_loss += loss.item()
    print(f"Epoch : {epoch+1} , Loss : {total_loss:.4f}")
```

```python
def predict(model,question,vocab,threshold=0.5,max_len=10):
    model.eval()
    index_to_word = {idx:word for word,idx in vocab.items()}

    with torch.no_grad():
        numberical_question = text_to_indices(question,vocab)
        
        # shape [1,seq_len]
        question_tensor = torch.tensor(numberical_question).unsqueeze(0)

        output = model(question_tensor,answer=None,max_len=max_len)

        # probabilities
        probs = torch.nn.functional.softmax(output,dim=2)
        value , index = torch.max(probs,dim=2)
        # print(f"Value :- {value},Index :- {index}")

        # answer in squance
        answer_words = []
        for i in range(index.size(1)):
            idx = index[0, i].item()
            confidence = value[0 , i].item()

            if idx == vocab.get('<EOS>',-1) or confidence < threshold:
                break

            answer_words.append(index_to_word.get(idx,'<UNK>'))
        
        answer = " ".join(answer_words)
        return f"Answer :- {answer}"
```
